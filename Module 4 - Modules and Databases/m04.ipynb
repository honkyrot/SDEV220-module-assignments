```python
#16.5 insert data into the books.db
the_csv_file_2 = 'books2.csv'
books_2 = []
import csv
with open(the_csv_file_2) as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        print(row)
        books_2.append(row)
```

```python
import sqlite3
conn = sqlite3.connect('books.db')
c = conn.cursor()

for book in books_2:
    c.execute("INSERT INTO books VALUES (?,?,?)",(book['title'],book['author'],int(book['year'])))
conn.commit()
conn.close()
```

```python
# 16.8 BACK AT THIS AGAIN
# use sqlalchemy to select and print the title column from the book table in alphabetical order.
import sqlalchemy
from sqlalchemy import text

engine = sqlalchemy.create_engine('sqlite:///books.db')
connection = engine.connect()
result = text("SELECT title FROM books ORDER BY title ASC")
result = connection.execute(result)
for row in result:
    print(row)
connection.close()


```
