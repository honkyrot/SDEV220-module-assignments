```
#9.2 define a generator function
def get_odds():
    odd_numbers = []
    for i in range(10):
        if i % 2 == 1:
            odd_numbers.append(i)
    
    return odd_numbers

get_odds()
```

```
for i, v in enumerate(get_odds()):
    if i == 2:
        print(v)
```
