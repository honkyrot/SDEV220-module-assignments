```python
class Solution:
    def sort012(self,arr,n):
        arr.sort()
        return arr
```

```python
the_class = Solution()
arr = [0, 2, 1, 2, 0]
n = 5
new_array = the_class.sort012(arr, n)
```

```python
# the binary search one
class Solution:	
    def binarysearch(self, arr:list, n:int, k:int):
        try:
            value = arr.index(k)
            return value
        except ValueError:
            return -1
```

```python
test = Solution()
print(test.binarysearch([1,2,3,4,5], 5, 6))
```

```python
test = Solution()
print(test.binarysearch([1,2,3,4,5], 5, 3))
```
